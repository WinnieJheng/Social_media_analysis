```python
import requests
import time
import json
from bs4 import BeautifulSoup
import csv

PTT_URL = 'https://www.ptt.cc'
```

```python
def get_web_page(url):
    resp = requests.get(
    url=url,
    cookies={'over18':'1'}
    )
    if resp.status_code != 200:
        print('Invalid url:',resp.url)
        return None
    else:
        return resp.text
```

```python
def get_articles(dom):
    soup = BeautifulSoup(dom, 'html5lib')

    # 取得上一頁的連結
    paging_div = soup.find('div', 'btn-group btn-group-paging')
    prev_url = paging_div.find_all('a')[1]['href']

    articles = []  # 儲存取得的文章資料
    divs = soup.find_all('div', 'r-ent')
    for d in divs:
        # 取得文章連結及標題
        if d.find('a'):  # 有超連結，表示文章存在，未被刪除
            href = d.find('a')['href']
            title = d.find('a').text
            articles.append({
                'title': title,
                'href': href,
                })
    return articles, prev_url
```

```python
current_page = get_web_page(PTT_URL + '/bbs/Hualien/index.html')
if current_page:
    articles = []  # 全部文章
        
    current_articles, prev_url = get_articles(current_page)  # 目前頁面的文章
    while current_articles:  # 若目前頁面有文章則加入 articles，並回到上一頁繼續尋找是否有文章
        articles += current_articles
        current_page = get_web_page(PTT_URL + prev_url)
        current_articles, prev_url = get_articles(current_page)
        if len(articles) > 10:
            break
            
    # 儲存或處理文章資訊
    print('共有', len(articles), '篇文章')
    for a in articles:
        print(a)
    with open('title_and_url.json', 'w', encoding='utf-8-sig') as f:
        
        json.dump(articles, f, indent=2, sort_keys=True, ensure_ascii=False)
```
